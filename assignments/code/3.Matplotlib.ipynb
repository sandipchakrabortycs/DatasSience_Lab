# Simple line plot

```python
import matplotlib.pyplot as plt
x = [60, 70, 80, 90]
y = [70, 80, 50, 85]
plt.plot(x, y)
plt.title("study time  vs marks")
plt.xlabel("duration of study")
plt.ylabel("Marks")
```

```python
import numpy as np
x = np.arange(1, 11)
y = x**2 + 4*x + 20
plt.plot(x, y)
print(x)
print(y)
plt.title("Samir")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.scatter(x, y)
plt.title("Simple Scatter Plot")
plt.show()
```

```python
plt.plot(x, y, color='green', marker='v', label='y = x^2')
plt.title("Customized Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(False)
plt.show()
```

#Subplot

```python
fig, axs = plt.subplots(3, 3)
axs[0, 0].plot(x, y)
axs[0, 1].scatter(x, y)
axs[1, 0].bar(['A', 'B', 'C'], [3, 7, 5])
axs[1, 1].hist([1, 2, 1, 2, 3, 4, 5])
axs[2, 0].pie([3, 7, 5])
axs[2, 1].boxplot([1, 2, 1, 2, 3, 4, 5])
axs[0, 2].plot(x, y)
axs[1, 2].scatter(x, y)
axs[2, 2].bar(['A', 'B', 'C'], [3, 7, 5])
plt.show()
plt.tight_layout()
plt.show()
plt.savefig("3_3.png", dpi=300, bbox_inches='tight')
```

```python
import seaborn as sns
import pandas as pd

# Sample data
data = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'A', 'B', 'C'],
    'Values': [1, 2, 3, 4, 5, 6]
})

# Bar plot
sns.barplot(x='Category', y='Values', data=data)
plt.title("Bar Plot Example")
plt.show()
```

```python
sns.boxplot(x='Category', y='Values', data=data)
plt.title("Box Plot Example")
plt.show()
```

```python
sns.set_palette("pastel")
sns.violinplot(x='Category', y='Values', data=data)
plt.title("Customized Violin Plot")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample DataFrame with missing values
iris = sns.load_dataset("iris")
df = pd.DataFrame(iris)

# Create a boolean mask of missing values
missing_mask = df.isnull()

# Plotting the missing values heatmap
plt.figure(figsize=(8, 4))
plt.imshow(missing_mask.T, aspect='auto', cmap='gray', interpolation='none')

# Customize the plot
plt.title('Missing Values Visualization')
plt.xlabel('Row Index')
plt.ylabel('Columns')
plt.yticks(ticks=range(df.shape[1]), labels=df.columns)
plt.colorbar(label='Missing (1) / Present (0)')

plt.show()
```
